```python
import pandas as pd
import numpy as np
import os

from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler 
from sklearn.metrics import accuracy_score
```

```python
#os.chdir("C:\Python Trainer\mlarmy\Online Class\PCA_Regularization")
wine_data = pd.read_csv("wine.csv")
wine_data.head()
wine_data.dtypes
```

```python
#Let's build DT before PCA 
X_bef_pca = wine_data.drop("wine.class", axis = 1)
y_bef_pca = wine_data["wine.class"]

#train test split
x_train, x_test, y_train, y_test = train_test_split(X_bef_pca, y_bef_pca, test_size = 0.3, random_state = 100)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)

#build DT
dt_bef_PCA = DecisionTreeClassifier().fit(x_train,y_train)
#acc on train
print("Acc on train: ", accuracy_score(y_train,dt_bef_PCA.predict(x_train)))
#acc on test
print("Acc on test: ", accuracy_score(y_test,dt_bef_PCA.predict(x_test)))
```

```python
#Lets do PCA 
X = wine_data.drop("wine.class", axis = 1)
y = wine_data["wine.class"]
np.set_printoptions(suppress = True)
pca_bef_sc = PCA(n_components=13).fit(X)
pca_bef_sc.explained_variance_ratio_
```

```python
#scale the data
X_new_np = StandardScaler().fit_transform(X)
X_new_df = pd.DataFrame(X_new_np, columns=X.columns)
X_new_df.head()
```

```python
#PCA on scaled data
pca_new = PCA().fit(X_new_df)
pca_new.explained_variance_ratio_
pca_new.components_
pca_new.explained_variance_
pca_new.inverse_transform(X_new_df)
print("Cumulative variance: ",np.cumsum(pca_new.explained_variance_ratio_))
pca_95 = PCA(n_components=0.95)
X_reduced = pca_95.fit_transform(X_new_df)
print(X_reduced.shape)
#convert into df
X_reduced_df = pd.DataFrame(X_reduced, columns= ["PCA"+ str(i) for i in range(1,X_reduced.shape[1]+1)])
X_reduced_df
```

```python
#build dt on the PCA df
#train test split
x_train, x_test, y_train, y_test = train_test_split(X_reduced_df, y_bef_pca, test_size = 0.3, random_state = 100)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)

#build DT
dt_bef_PCA = DecisionTreeClassifier().fit(x_train,y_train)
#acc on train
print("Acc on train: ", accuracy_score(y_train,dt_bef_PCA.predict(x_train)))
#acc on test
print("Acc on test: ", accuracy_score(y_test,dt_bef_PCA.predict(x_test)))
```
